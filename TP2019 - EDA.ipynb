```python
pesajes = pd.read_csv(r'pesajes_nulos_limpio.csv', delimiter=';')
```

```python
pesajes.head()
```

```python
sns.set()
cols = ['VIDRIO', 'PAPEL_Y_CARTON', 'METAL', 'TELGOPOR', 'PLASTICOS_TODOS_LOS_DEMAS', 'TETRA_BRICK']
sns.pairplot(pesajes[cols], size = 2.5)
plt.show()
```

Con el scatter plot podemos observar que tanto telgopor, metal y tetrabrick no hay cantidades imporantes, salvo 2 o 3 casos aislados. No asi en plasticos, papel y vidrio, en donde se manejan grandes cantidades de kilos, plasticos con un promedio de 150 kilos, y tanto papel como vidrio con promedios de 250 kilos.

# Analisis de cada material

Ahora analizaremos caso por caso cada material para evaluar cantidad recogida

### Vidrio

```python
ay = sns.boxplot(x = pesajes.VIDRIO, y = pesajes.PUNTO, data = pesajes)
ay.set_xticklabels(ay.get_xticklabels(), rotation=90, ha="right")
plt.title("distribucion de vidrio por semana en los puntos verdes")
plt.show()
```
