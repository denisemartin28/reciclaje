# Analisis cantidad poblacion por comuna

En este punto vamos a analizar la relacion entre la densidad de poblacion de la comuna y la cantidad reciclada, para ver si podemos encontrar una relacion. Utilizaremos los datos del censo 2010, ya que fue el ultimo censo nacional realizado en el pais (https://data.buenosaires.gob.ar/dataset/informacion-censal-por-radio).

```python
poblacion = pd.read_csv(r'informacion-censal-por-radio-2010.csv', delimiter=';')
```

```python
poblacion_comuna = poblacion.groupby(["COMUNA"]).sum()
poblacion_comuna.head()
```

```python
poblacion_comuna_total = poblacion_comuna.sort_values(by='TOTAL_POB', ascending=False)
poblacion_comuna_total
```

Las comunas con mayor poblacion como la 13, 14, 7, 4 y 1 no son las comunas que se destaquen puntualmente por ser las que mas reciclan, como tampoco la comuna con menor poblacion vemos alguna relacion, la comuna 9 es una de las menores con poblacion y la que segunda menos recicla, pero inmediatamente encima tiene a la comuna 10, con solo 5000 habitantes mas, y es la que mas recicla de todas, por lo que vemos que tampoco hay correlacion en este punto.

Aprovechando que el archivo del censo 2010 encontrado tiene una columna con hogares con necesidades basicas insatisfechas, vamos a analizarla desde ese punto para buscar alguna relacion.

```python
poblacion_comuna_total = poblacion_comuna.sort_values(by='H_CON_NBI', ascending=False)
poblacion_comuna_total
```

En este punto empezamos a notar un indicio de lo estudiado. Entre las 6 comunas con menos necesidades basicas insatisfechas, vemos las 5 que mas reciclan se encuentran en este punto (comuna 6, 10, 11, 12 y 13), mientras que entre las 5 que mas hogares con necesidades basicas insatisfechas hay, se encuentran 3 comunas que son las que menos reciclan. Como este numero solo aisladamente no nos dice nada, vamos a sacar el porcentaje de hogares con necesidades basicas insatisfechas por el total de hogares por comuna, para ver si este analisis se mantiene

```python
NBI = poblacion_comuna_total.iloc[:,9:12]
NBI
```

```python
percent_NBI = (NBI.H_CON_NBI/NBI.T_HOGAR).sort_values(ascending=False)
percent_NBI
```

Viendo que se mantiene la tendencia, las comunas donde menos de un 3,5% de los hogares tienen necesidades basicas insatisfechas son las que mas reciclan, mientras que las comunas con mas de 11% de necesidades basicas insatisfechas en sus hogares se encuentran comuna con un porcentaje de poco reciclaje en peso, por lo que podemos sacar conclusiones de estos puntos.

El INDEC considera como hogares con necesidades basicas insatisfechas los siguientes tipos:

- NBI 1. Viviendas: vivienda que habitan los hogares que moran en habitaciones de inquilinato, hotel o pensión, viviendas no destinadas a fines habitacionales, viviendas precarias y otro tipo de vivienda. Se excluye a las viviendas tipo casa, departamento y rancho.
- NBI 2. Condiciones sanitarias: incluye a los hogares que no poseen retrete.
- NBI 3. Hacinamiento: es la relación entre la cantidad total de miembros del hogar y la cantidad de habitaciones de uso exclusivo del hogar. Operacionalmente se considera que existe hacinamiento crítico cuando en el hogar hay más de tres personas por cuarto.
- NBI 4. Asistencia escolar: hogares que tienen al menos un niño en edad escolar (6 a 12 años) que no asiste a la escuela.
- NBI 5. Capacidad de subsistencia: incluye a los hogares que tienen cuatro o más personas por miembro ocupado y que tienen un jefe no ha completado el tercer grado de escolaridad primaria.
